```python
import pandas as pd
import numpy as np
import scipy.stats as s
```

```python
raw_data = pd.read_csv("data.csv")
```

```python
raw_data.drop([cols[0],cols[32]],axis=1,inplace=True)
```

```python
raw_data.head()
```

```python
benign_tumors_data = pd.DataFrame(raw_data[raw_data['diagnosis'] == 'B'])
```

```python
benign_tumors_data.drop(['diagnosis'],axis=1,inplace=True)

benign_tumor_data_array = np.array(benign_tumors_data)
```

```python
alpha = 0.001
epsilon = 0.00001
initial_vals = np.random.randn(2,1)

while(True):
    
    del_mucap = -((np.sum((benign_tumor_data_array[:,0]-initial_vals[0])))/initial_vals[1]**2)
    
    del_sigmacap = ((benign_tumor_data_array.shape[0]/initial_vals[1]) - ((np.sum((benign_tumor_data_array[:,0]-initial_vals[0])**2))/initial_vals[1]**3))
    
    dels = np.array([del_mucap,del_sigmacap])
    
    dels = dels.reshape(2,1)
    
    final_vals = initial_vals - (alpha * dels)
    
    func_old_val = -np.sum(np.log(s.norm.pdf(benign_tumor_data_array[:,0],initial_vals[0],initial_vals[1])))
    
    func_new_val = -np.sum(np.log(s.norm.pdf(benign_tumor_data_array[:,0],final_vals[0],final_vals[1])))
    
    if abs(func_old_val - func_new_val) < epsilon:
        
        break
        
    print("The initial and final value of Neg Log Likelihood is {} and {}".format(func_old_val,func_new_val))
    
    initial_vals = final_vals
```
